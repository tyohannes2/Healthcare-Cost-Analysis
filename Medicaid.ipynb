```python
import pandas as pd

# Path to your uploaded CSV file
file_path = "/content/drive/MyDrive/MUP_PHY_R25_P05_V20_D23_Prov_Svc.csv"

# Load the CSV into a DataFrame
df = pd.read_csv(file_path, low_memory=False)

# Output shape and first few rows
print("Shape of dataset:", df.shape)
df.head()
```

```python
# Check the column names
print(df.columns.tolist())

# See data types and missing values
df.info()

# Look at summary statistics
df.describe()
```

```python
# Group by state and calculate the average Medicare payment
state_avg = df.groupby('Rndrng_Prvdr_State_Abrvtn')['Avg_Mdcr_Pymt_Amt'].mean().reset_index()

# Sort from highest to lowest
state_avg = state_avg.sort_values(by='Avg_Mdcr_Pymt_Amt', ascending=False)

# Display top 10 states
state_avg.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.bar(state_avg['Rndrng_Prvdr_State_Abrvtn'], state_avg['Avg_Mdcr_Pymt_Amt'], color='skyblue')
plt.title('Average Medicare Payment Amount by State (2023)', fontsize=14)
plt.xlabel('State')
plt.ylabel('Avg Medicare Payment ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
us_states = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
    'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD',
    'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ',
    'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC',
    'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY', 'DC'
]
```

```python
# Filtering the DataFrame to include only 50 states (and DC)
df_us = df[df['Rndrng_Prvdr_State_Abrvtn'].isin(us_states)]
```

```python
state_avg_us = df_us.groupby('Rndrng_Prvdr_State_Abrvtn')['Avg_Mdcr_Pymt_Amt'].mean().reset_index()
state_avg_us = state_avg_us.sort_values(by='Avg_Mdcr_Pymt_Amt', ascending=False)

# View top 10
state_avg_us.head(10)
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(15, 6))

# Create bar chart
plt.bar(state_avg_us['Rndrng_Prvdr_State_Abrvtn'], state_avg_us['Avg_Mdcr_Pymt_Amt'], color='steelblue')

# Add labels and title
plt.title('Average Medicare Payment by State (2023)', fontsize=16)
plt.xlabel('State')
plt.ylabel('Average Medicare Payment ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the chart
plt.tight_layout()
plt.show()
```
